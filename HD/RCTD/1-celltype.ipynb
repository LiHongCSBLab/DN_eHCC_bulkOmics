### define cell type

```python
adata.obs['celltype'] = 'Unassigned'
tmp_df = pd.DataFrame(norm_weights.idxmax(axis=1))

for i in tmp_df.index:
    adata.obs.loc[i,'celltype'] = tmp_df.loc[i,0]
adata.obs['celltype'] = adata.obs['celltype'].replace('Tumor', 'Hepatocyte')
```

### group

```python
# adata = sc.read('/csb2/finishedProject/HuiLiJian_DN_2025/10x_HD/5-final/2-RCTD/178248A.016um/178248A.016um.celltype.h5ad')
df = pd.read_csv('/csb2/finishedProject/HuiLiJian_DN_2025/10x_HD/5-final/0-Rawh5ad/S178248A.016um.圈画.csv',index_col=0)
```
