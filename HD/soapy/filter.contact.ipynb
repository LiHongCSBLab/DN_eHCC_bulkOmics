## 1

```python
samples = ['17s30980_DN','18s18941_DN','178248A_DN','18s18941_t2_DN',
          '17s30980_Tumor','18s18941_Tumor','178248A_Tumor','18s18941_t2_Tumor']
savepath = '/csb2/finishedProject/HuiLiJian_DN_2025/10x_HD/5-final/test/SOAPy/CCC_csv/'
```

```python
output = {}
```

```python
for sample in samples:

    df1 = pd.read_csv(savepath+f'{sample}_contact_affinity.csv',index_col=0)
    df2 = pd.read_csv(savepath+f'{sample}_contact_strength.csv',index_col=0)
    # df3 = pd.read_csv(savepath+f'{sample}_secretory_affinity.csv',index_col=0)
    # df4 = pd.read_csv(savepath+f'{sample}_secretory_strength.csv',index_col=0)

    tmp1 = df2[(df1 <0.05) & (df2 >2)]
    # tmp2 = df4[(df3 <0.05) & (df4 >2)]
    res1 = [f"{row},{col}" for row, col in tmp1.stack().index.tolist()]
    # res2 = [f"{row},{col}" for row, col in tmp2.stack().index.tolist()]
    output[sample] = res1
```

```python
df = pd.DataFrame(list(output.items()), columns=['sample', 'LR'])
df.to_csv('/csb2/finishedProject/HuiLiJian_DN_2025/10x_HD/5-final/test/SOAPy/CCC_csv/filter/CCC.contact.strength2.p0.05.csv', index=False)
```

```python
output = {}

for sample in samples:

    # df1 = pd.read_csv(savepath+f'{sample}_contact_affinity.csv',index_col=0)
    # df2 = pd.read_csv(savepath+f'{sample}_contact_strength.csv',index_col=0)
    df3 = pd.read_csv(savepath+f'{sample}_secretory_affinity.csv',index_col=0)
    df4 = pd.read_csv(savepath+f'{sample}_secretory_strength.csv',index_col=0)

    # tmp1 = df2[(df1 <0.05) & (df2 >2)]
    tmp2 = df4[(df3 <0.05) & (df4 >2)]
    # res1 = [f"{row},{col}" for row, col in tmp1.stack().index.tolist()]
    res2 = [f"{row},{col}" for row, col in tmp2.stack().index.tolist()]
    output[sample] = res2

df = pd.DataFrame(list(output.items()), columns=['sample', 'LR'])
df.to_csv('/csb2/finishedProject/HuiLiJian_DN_2025/10x_HD/5-final/test/SOAPy/CCC_csv/filter/CCC.secretory.strength2.p0.05.csv', index=False)
```

```python
df = pd.read_csv('/csb2/finishedProject/HuiLiJian_DN_2025/10x_HD/5-final/test/SOAPy/CCC_csv/filter/CCC.strength2.p0.05.csv',index_col=0)
from collections import Counter
import ast

# 方法A：使用 apply
df['LR'] = df['LR'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
all_pairs_dict = {}
for idx in df.index:
    lr_list = df.loc[idx, 'LR']
    cell_pairs = [item.split(',')[1] for item in lr_list]
    count_result = Counter(cell_pairs)
    
    for pair, count in count_result.items():
        if pair not in all_pairs_dict:
            all_pairs_dict[pair] = {}
        all_pairs_dict[pair][idx] = count

result_df = pd.DataFrame.from_dict(all_pairs_dict, orient='index').fillna(0).astype(int)

# result_df.to_csv('/csb2/finishedProject/HuiLiJian_DN_2025/10x_HD/5-final/test/SOAPy/CCC_csv/filter/CCC.contact.strength2.p0.05.N.csv')
```
